```python
def highest_rated_ids(N, arr):
    ratings_dict = {}

    for i in range(N):
        ID, rating = arr[i][0], arr[i][1]
        if ID not in ratings_dict or rating > ratings_dict[ID][0] or (rating == ratings_dict[ID][0] and i < ratings_dict[ID][1]):
            ratings_dict[ID] = (rating, i)

 
    highest_rating = max(ratings_dict.values(), key=lambda x: x[0])[0]

  
    highest_rated_ids = [ID for ID, (rating, _) in ratings_dict.items() if rating == highest_rating]

    highest_rated_ids.sort()

    return highest_rated_ids[0]

# Example usage:
N = 5
observations = [
    [512, 4.5],
    [123, 3.8],
    [100, 4.5],
    [757, 4.2],
    [987, 3.8]
]

result = highest_rated_ids(N, observations)
print("IDs with the highest rating:", result)
```
